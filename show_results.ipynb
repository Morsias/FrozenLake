```python
CHECKPOINT_PATH = "models/PPO_FrozenLake8x8-v1_2021-08-29_12-44-17hwu4pe41/checkpoint_000210/checkpoint-210"
TRAINING_RESULTS_DQN = "results/training_results_DQN_2021_08_31_09_29_14.csv"
TRAINING_RESULTS_PPO = "results/training_results_PPO_2021_08_31_09_37_06.csv"
INFERENCE_RESULTS = "results/inference_results.csv"
```

```python
%matplotlib inline
import ray
import ray.rllib.agents.ppo as ppo
import gym
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

```

```python
infer_results = pd.read_csv(INFERENCE_RESULTS, index_col=0)
dqn_train_results = pd.read_csv(TRAINING_RESULTS_DQN, index_col=0)
ppo_train_results = pd.read_csv(TRAINING_RESULTS_PPO, index_col=0)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
fig.suptitle('Training Results')
sns.lineplot(ax=axes[0], x=dqn_train_results.index, y="MeanEpisodeReward",data=dqn_train_results)
sns.lineplot(ax=axes[1], x=ppo_train_results.index, y="MeanEpisodeReward",data=ppo_train_results)
axes[0].set_title("DQN")
axes[0].set(xlabel='Episode Iteration', ylabel='Mean Reward')
axes[0].set_ylim(0,1)
axes[1].set_title("PPO")
axes[1].set(xlabel='Episode Iteration', ylabel='Mean Reward')
axes[1].set_ylim(0,1)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
fig.suptitle('Inference Results')
sns.countplot(ax=axes[0], data=infer_results, x='EpisodeReward')
sns.countplot(ax=axes[1], data=infer_results, x='EpisodeReward')
axes[0].set_title("DQN")
axes[0].set(xlabel='Episode Reward', ylabel='Number of Episodes')
axes[0].set_ylim(0,1000)
axes[1].set_title("PPO")
axes[1].set(xlabel='Episode Reward', ylabel='Number of Episodes')
axes[0].set_ylim(0,1000)
```

Import model from checkpoint

```python
# Setup environment
env = gym.make("FrozenLake8x8-v1")
sns.set_theme(style="darkgrid")

ray.init()
config = ppo.DEFAULT_CONFIG.copy()
# Initialize agent
agent = ppo.PPOTrainer(config=config, env="FrozenLake8x8-v1")
# Restore from checkpoint
agent.restore(CHECKPOINT_PATH)
```

Run rendered example

```python
observation = env.reset()
done = False
# Run episode until it is over
while not done:
    env.render()
    # Compute action from agent
    # explore=False to ensure that no exploration takes place during inference
    action = agent.compute_single_action(observation, explore=False)
    observation, reward, done, info = env.step(action)
```

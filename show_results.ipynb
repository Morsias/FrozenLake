```python
CHECKPOINT_PATH = "models/PPO_FrozenLake8x8-v1_2021-08-29_12-44-17hwu4pe41/checkpoint_000210/checkpoint-210"
```

```python
%matplotlib inline
import ray
import ray.rllib.agents.ppo as ppo
import gym
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Setup environment
env = gym.make("FrozenLake8x8-v1")
sns.set_theme(style="darkgrid")
```

```python
run_results = pd.read_csv("results/inference_results.csv", index_col=0)
```

```python
sns.countplot(x='EpisodeReward', data=run_results)
```

Import model from checkpoint

```python
ray.init()
config = ppo.DEFAULT_CONFIG.copy()
# Initialize agent
agent = ppo.PPOTrainer(config=config, env="FrozenLake8x8-v1")
# Restore from checkpoint
agent.restore(CHECKPOINT_PATH)
```

Run rendered example

```python
observation = env.reset()
done = False
# Run episode until it is over
while not done:
    env.render()
    # Compute action from agent
    # explore=False to ensure that no exploration takes place during inference
    action = agent.compute_single_action(observation, explore=False)
    observation, reward, done, info = env.step(action)
```

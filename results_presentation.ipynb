## Frozen lake results
In this notebook the results of training and inference for the FrozeLake environment are presented.
The algorithms used were PPO and DQN

The final cell runs a rendered example of the optimal solution found by the algorithms


```python
CHECKPOINT_PATH = "models/PPO_2021_08_31_09_37_06/checkpoint_000200/checkpoint-200"
TRAINING_RESULTS_DQN = "results/training_results_DQN_2021_08_31_09_29_14.csv"
TRAINING_RESULTS_PPO = "results/training_results_PPO_2021_08_31_09_37_06.csv"
INFERENCE_RESULTS_DQN = "results/inference_results_DQN_2021_08_31_21_39_08.csv"
INFERENCE_RESULTS_PPO = "results/inference_results_PPO_2021_08_31_21_41_44.csv"
```

```python
%matplotlib inline
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from inference import inference_model
```

```python
dqn_infer_results = pd.read_csv(INFERENCE_RESULTS_DQN, index_col=0)
ppo_infer_results = pd.read_csv(INFERENCE_RESULTS_PPO, index_col=0)
dqn_train_results = pd.read_csv(TRAINING_RESULTS_DQN, index_col=0)
ppo_train_results = pd.read_csv(TRAINING_RESULTS_PPO, index_col=0)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
fig.suptitle('Training Results')
sns.lineplot(ax=axes[0], x=dqn_train_results.index, y="MeanEpisodeReward",data=dqn_train_results)
sns.lineplot(ax=axes[1], x=ppo_train_results.index, y="MeanEpisodeReward",data=ppo_train_results)
axes[0].set_title("DQN")
axes[0].set(xlabel='Episode Iteration', ylabel='Mean Reward')
axes[0].set_ylim(0,1)
axes[1].set_title("PPO")
axes[1].set(xlabel='Episode Iteration', ylabel='Mean Reward')
axes[1].set_ylim(0,1)
```

DQN is "slower" in achieving higher mean rewards. This is due to high amounts of exploration
given in the DQN configuration.


```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
fig.suptitle('Inference Results')
sns.countplot(ax=axes[0], data=dqn_infer_results, x='EpisodeReward')
sns.countplot(ax=axes[1], data=ppo_infer_results, x='EpisodeReward')
axes[0].set_title("DQN")
axes[0].set(xlabel='Episode Reward', ylabel='Number of Episodes')
axes[0].set_ylim(0,1000)
axes[1].set_title("PPO")
axes[1].set(xlabel='Episode Reward', ylabel='Number of Episodes')
axes[1].set_ylim(0,1000)
```

The results are practically the same. (867 successful episodes for DQN, 866 for PPO).
